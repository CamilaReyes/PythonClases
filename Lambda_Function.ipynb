# Mapping or map () function

```python
# en el primer argumento recibe una funcions, esta se aplica a la lista o dupla que le pasemos de datos
x=[1,2,3,4,5,6]

suaredElements= map(lambda x : x**2, x) # esta ultima x es la lista

print(list(suaredElements))
```

```python
def tempe(T):
    return (((float(9)/5))*(T-275) + 32.0)

def celsios(T):
    return T-275
temperaturas=[36.5,37,37.5,38,39]

F= list(map(tempe, temperaturas))
C= list(map(celsios, F))

print(F)
print(C)
```

```python
tempe2= lambda z : ((float(9)/5))*z + 32.0

tempe3= lambda z : (5.0/9.0)*(z-32)


temperaturas2=[36.5,37,37.5,38,39]

A= list(map(tempe2, temperaturas2))
B= list(map(tempe3, A))

print(A)
print(B)
```

# Reducing

*import functools
*reduce(funcion, seq)

```python
from functools import reduce


reduce(lambda x, y : x+y, [47,11,42,13]) #suma los valores entre ellos, 47+11=58, 58+42=100....
```

```python
num=[1,6,545,5464536,546456457]
max(num)
```

```python
reduce(lambda x,y: x if (x>y) else y, num) #sacar el numero mayor de la lista. que compare entre x y y cual es el mayor, si es x que devuleva x sino que devueva y
```

```python
avengers=["captain america", "iron man", "spiderman", "hulk", "thor"]

reduce(lambda x,y: x+y, avengers)
```
