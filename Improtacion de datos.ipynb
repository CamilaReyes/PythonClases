```python
import pickle
```

```python
numberData= int(input("Enter length of list: "))
data=[]

for i in range (numberData):
    raw= input("Enter Data" + str(i) + " : ")
    data.append(raw)

file= open('./Pfilfe', 'wb')
pickle.dump(data, file)

file.close()
```

```python
with open('./Pfilfe', 'rb') as f:
    dataLectura= pickle.load(f)
```

```python
cnt= 0
for item in dataLectura:
    print("The data", cnt, "is :", item)
    cnt+=1
```

```python
#Ir a una hoja en especifico
print(type(xls))
```

```python
print(xls.parse('2004'))#Ir a una hoja en especifico
```

```python
print(xls.parse(0).head()) #Llamar una hoja en especifico
```
